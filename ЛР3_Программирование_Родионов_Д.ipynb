# Обзор модуля ***collections***

## Counter

***collections.Counter*** - вид словаря, который позволяет нам считать количество неизменяемых объектов (в большинстве случаев, строк).

```python
from collections import Counter

c = Counter()
for word in ['one', 'two', 'two', 'three', 'three', 'three']:
  c[word] += 1

print(c)
print(c['three'])
```

## Deque

***collections.deque(iterable, [maxlen])*** - создаёт очередь из итерируемого объекта с максимальной длиной maxlen. Очереди очень похожи на списки, за исключением того, что добавлять и удалять элементы можно либо справа, либо слева.

```python
from collections import deque

d = deque()
d.append(2)
d.append(3)
d.appendleft(1)

print(d)
d.reverse()
print(d)
```

## defaultdict

***collections.defaultdict*** ничем не отличается от обычного словаря за исключением того, что по умолчанию всегда вызывается функция, возвращающая значение

## OrderedDict

***collections.OrderedDict*** - ещё один похожий на словарь объект, но он помнит порядок, в котором ему были даны ключи. 

## namedtuple

Класс ***collections.namedtuple*** позволяет создать тип данных, ведущий себя как кортеж, с тем дополнением, что каждому элементу присваивается имя, по которому можно в дальнейшем получать доступ

```python
# Пример с версией 3.7

from collections import namedtuple

nt = namedtuple("Name", ["Field1", "Field2"])
n = nt(1, 2)

print(nt)
```

## ChainMap()

Класс ***ChainMap()*** предназначен для быстрого объединения нескольких словарей, чтобы их можно было рассматривать как единое целое. Такой контейнер объединяет словари и ищет ключи намного быстрее, чем создание нового словаря и выполнение объединения при помощи вызовов dict.update().

```python
from collections import ChainMap

f = {'two': 2, 'three': 3}
s = {'one': 1, 'two': 2}
res = ChainMap(f, s)

print(res)
```

## UserDict, UserList, UserString

Классы ***UserDict, UserList, UserString*** действует как класс-оболочка для объектов словаря, списка и строк соответственно. Эти классы полезны, когда нужно создать собственный словарь, список, строку с некоторыми измененными функциями или с некоторыми новыми функциями. Это можно рассматривать как способ добавления нового поведения.

#Обзор версии 3.8

Метод _asdict() из collections.namedtuple() теперь возвращает словарь (dict) вместо orderedDict 

```python
# Пример с версией 3.7

from collections import namedtuple

nt = namedtuple("Name", ["Field1", "Field2"])
n = nt(1, 2)

d = n._asdict()
print(type(d))
```
